```python
# Function: Block of code which performs a specific task
# Characteristics:
# 1. Reusable: Can be called multiple times
# 2. Modular: Helps in breaking down complex problems
# 3. Organized: Improves code readability and maintainability
# Defining a function using 'def' keyword
# Function Name: Identifier for the function
# Parameters: Inputs to the function (optional)
# Types of Functions: 
# 1. Built-in Functions (e.g., print(), len(), type())
# 2. User-defined Functions (defined by the programmer)
# Syntax:
# def function_name():
#     # code block
# Example:
def greet():
    print("Hello, welcome to the world of Functions!")  
greet()
# This is a simple function that prints a greeting message.

def sum (a,b):
    return a+b
res = sum(5,10)
print(f"Sum of 5 and 10 is: {res}")
```

```python
def even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
result = even_odd(7)
print(f"The number 7 is: {result}")

# Default Parameters
def power(base, exponent=2):
    return base ** exponent
result = power(4)
print(f"4 raised to the power of 2 is: {result}")   
result = power(3, 3)
print(f"3 raised to the power of 3 is: {result}")
```

```python
# Variable Length Arguments
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result
product = multiply(2, 3, 4)
print(f"Product of 2, 3, and 4 is: {product}")
# Keyword Arguments
def introduce(name, age):
    print(f"My name is {name} and I am {age} years old.")
introduce(age=25, name="Alice")
```

```python
# normal > * args > default > **kwargs
def func(a, b, *args, c=10, **kwargs):
    print(f"a: {a}, b: {b}")
    print(f"args: {args}")
    print(f"c: {c}")
    print(f"kwargs: {kwargs}")
func(1, 2, 3, 4, 5, c=20, d=30, e=40) 
```
