```python
a = int(input("Enter the No 1 : " ))
b = int(input("Enter the No 2 : " ))
print("Sum of two numbers is : ", a + b)
```

```python
n = int(input("Enter number of rows: "))
for i in range(1, n + 1):
    print(" " * (n - i), end="")
    print("*" * (2 * i - 1))

```

```python
n = int(input())
a = [list(map(int, input().split())) for _ in range(n)]
for i in range(n):
    a[i][i] = 0
for r in a:
    
    print(*r)
```

```python
n = int(input())

arr = []
for i in range(n):
    arr.append(list(map(int, input().split())))

# change diagonal
for i in range(n):
    arr[i][i] = 0

# output
for i in range(n):
    for j in range(n):
        print(arr[i][j], end=" ")
    print()
```
