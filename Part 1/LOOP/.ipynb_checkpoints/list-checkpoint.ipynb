```python
#FRUITS LIST
fruits = ["apple","gauave","banana","cherry","orange","kiwi","melon","mango"]
#slicing
print(fruits[1:3])# banana to cherry
print(fruits[:4])# apple to cherry
print(fruits[2:])# banana to end
print(fruits[-4:-1])# orange to melon
#reverse slicing
print(fruits[::-1])
```

```python
fruits = ["apple","banana","cherry"]
#modifying elements
fruits[0] = "kiwi"
print(fruits)
fruits[1:] = "mango"
print(fruits)
```

```python
#list methods
fruits = ["apple","banana","cherry"]
fruits.append("orange") #adds at the end
print(fruits)
fruits.insert(1,"lemon") #adds at specific index
print(fruits)
fruits.remove("banana") #removes specific element
print(fruits)   
fruits.pop() #removes last element
print(fruits)
fruits.pop(1) #removes element at specific index
print(fruits)
fruits.clear() #clears the list
print(fruits)
fruits = ["apple","banana","cherry","orange","kiwi","melon","mango"]
print(fruits.index("cherry")) #returns index of specific element
print(fruits.count("apple")) #returns count of specific element
fruits.sort() #sorts the list
print(fruits)
```

```python
#slicing methods
#            -7      -6      -5       -4      -3       -2    -1   
fruits = ["apple","banana","cherry","orange","kiwi","melon","mango"]
           # 0       1        2        3        4       5   
           #      6
print(fruits[2:5]) #cherry to kiwi
print(fruits[:4])  #apple to orange
print(fruits[3:])  #orange to end
print(fruits[-4:-1]) #orange to melon
```

```python
#iterating through list using while loop
fruits = ["apple","banana","cherry","orange","kiwi","melon","mango"]
i = 0
while i < len(fruits):
    print(fruits[i])
    i += 1
# --- IGNORE ---
# End of recent edits
```
