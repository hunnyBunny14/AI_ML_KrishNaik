```python
str = "Hello"
for char in str:
    print(char)
#string is iterable object and collections are iterable objects
#while loop
count = 0
while count < 5:
    print(count)
    count += 1  

list = [2,4,44,8,16,32]
i = 0
while i < len(list) and list[i] % 2 == 0:
    print(list[i])
    i += 1
```

```python
# use of pass and continue and break
for i in range(10):
    if i == 5:
        pass  # placeholder, does nothing
    print(f"passed iteration {i}")
for i in range(10):
    if i % 2 == 0:
        continue  # skip even numbers
    print(f"odd number: {i}")
for i in range(10):
    if i == 5:
        break  # exit loop when i is 5
    print(f"number: {i}")
```

```python
#nested loops
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")
```

```python
#calculate the sum of first n natural numbers with for loop
n = 10
sum_n = 0
for i in range(1, n + 1):
    sum_n += i
print(f"Sum of first {n} natural numbers is: {sum_n}")
#calculate the sum of first n natural numbers with while loop
n = 10
sum_n = 0
i = 1
while i <= n:
    sum_n += i
    i += 1
print(f"Sum of first {n} natural numbers is: {sum_n}")
```

```python
#1 to 100 prime no
for i in range(101):
    if i>1:
        for c in range(2,i):
            if(i%c==0):
                break
        else:
            print(i)
```
