```python
print("Lets Study Tuples Data Structure in Python")
# A tuple is an immutable sequence type in Python, 
# meaning once it is created, its elements cannot be changed.
# Tuples are defined by enclosing elements in parentheses ().   
# Example of creating a tuple
my_tuple = (1, 2, 3, "Hello", True)
print("My Tuple:", my_tuple)
# Accessing elements in a tuple
print("First Element:", my_tuple[0])
print("Last Element:", my_tuple[-1])
# Tuples can contain elements of different data types
mixed_tuple = (42, "Python", 3.14, False)
print("Mixed Tuple:", mixed_tuple)
# Tuples support indexing and slicing
print("Sliced Tuple (1:4):", my_tuple[1:4])
# Tuples are immutable, so we cannot change their elements
# The following line would raise an error if uncommented
# my_tuple[0] = 10
```
