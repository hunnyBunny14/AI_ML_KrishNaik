```python
# Import the built-in math module from Python's standard library
import math

# math.sqrt() is used to calculate the square root of a number
# Here, it calculates the square root of 16
print(math.sqrt(16))  # Output will be 4.0
```

```python
# RANDOM: It gives random values based on the given range or collection
import random

# Generates a random integer between 1 and 10 (both included)
print(random.randint(1, 10))

# Selects one random character from the given string
print(random.choice('ABCD'))

# Selects one random element from the given list
print(random.choice(['A', 'B', 'C', 'D']))
```

```python
# Import the os module from Python standard library
import os

# Prints the current working directory (where your Python file is running)
print(os.getcwd())

# Prints the list of files and folders in the current directory
print(os.listdir())

# Prints the name of the operating system
# 'nt'  → Windows
# 'posix' → Linux / macOS
print(os.name)

# Prints all environment variables of the system
# (stored as key-value pairs)
print(os.environ)
```

```python
# Import json module from Python standard library
import json

# Create a Python dictionary
data = {
    "name": "Sahil",
    "age": 20
}

# Convert Python dictionary to JSON string
json_data = json.dumps(data)

# Print the JSON formatted string
print(json_data)
```
